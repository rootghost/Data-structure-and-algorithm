```python
def max_pairwise_product_naive(y):
    max_product = 0
    for i in range(0,len(y)):
        for j in range(i+1,len(y)):
            product = y[i]*y[j]
            if product>max_product:
                max_product = product
    return max_product
            
x = input().split()
a = []
for i in range(0,len(x)):
    a.append(int(x[i]))
print(max_pairwise_product_naive(a))
```
